```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
```

```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
import matplotlib.pyplot as plt

arrow_prop_dict = dict(mutation_scale=10, shrinkA=0, shrinkB=0, linewidth=1.5)


class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)


zheight = 0.4
points = np.array([[-1, -1, -zheight],
                  [1, -1, -zheight ],
                  [1, 1, -zheight],
                  [-1, 1, -zheight],
                  [-1, -1, zheight],
                  [1, -1, zheight ],
                  [1, 1, zheight],
                  [-1, 1, zheight]])
Z = points
Z = Z
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111, projection='3d')
r = [-1,1]
X, Y = np.meshgrid(r, r)
verts = [[Z[0],Z[1],Z[2],Z[3]],
 [Z[4],Z[5],Z[6],Z[7]],
 [Z[0],Z[1],Z[5],Z[4]],
 [Z[2],Z[3],Z[7],Z[6]],
 [Z[1],Z[2],Z[6],Z[5]],
 [Z[4],Z[7],Z[3],Z[0]]]
ax.add_collection3d(Poly3DCollection(verts, facecolors=None, linewidths=0.5, edgecolors='b', alpha=0.1, linestyle='--'))
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
ax.axes.zaxis.set_ticklabels([])
for line in ax.xaxis.get_ticklines():
    line.set_visible(False)
for line in ax.yaxis.get_ticklines():
    line.set_visible(False)
for line in ax.zaxis.get_ticklines():
    line.set_visible(False)
ax.set_zlim(-1,1)
ax.set_axis_off()
gridpoints = np.linspace(-0.8,0.8,5)
nvs = np.array([(x,y, 0) for x in gridpoints for y in gridpoints])
ax.scatter(nvs[:,0],nvs[:,1],nvs[:,2]-0.05, label='NV-centre', c='r', alpha=1, s=100)
for (x,y,z) in nvs:
    arrow = Arrow3D([x+0.018, x+0.018],[y-0.1,y+0.1],[z-0.1,z+0.1], arrowstyle='-|>', color='k', **arrow_prop_dict)
    ax.add_artist(arrow)

ax.legend(loc=4)

nv1, nv2 = (gridpoints[1],gridpoints[0],0), (gridpoints[2],gridpoints[0],0)
arrow = Arrow3D([nv1[0],nv2[0]],[nv1[1],nv2[1]], [nv1[2]-0.05,nv2[2]-0.05],arrowstyle='<|-|>', color='b', mutation_scale=20)
ax.add_artist(arrow)
ax.text(np.mean([nv1[0],nv2[0]])-0.1,np.mean([nv1[1],nv2[1]]), np.mean([nv1[2]-0.05,nv2[2]-0.05]),"7nm", fontdict={"fontsize":"large"})
ax.xaxis.set_major_locator(plt.NullLocator())
ax.yaxis.set_major_locator(plt.NullLocator())
ax.zaxis.set_major_locator(plt.NullLocator())
plt.savefig('regular.pdf', bbox_inches='tight', transparent=True)
plt.show()
```

```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
import matplotlib.pyplot as plt

arrow_prop_dict = dict(mutation_scale=10, shrinkA=0, shrinkB=0, linewidth=1.5)


class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)



zheight = 0.4
points = np.array([[-1, -1, -zheight],
                  [1, -1, -zheight ],
                  [1, 1, -zheight],
                  [-1, 1, -zheight],
                  [-1, -1, zheight],
                  [1, -1, zheight ],
                  [1, 1, zheight],
                  [-1, 1, zheight]])

Z = points
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111, projection='3d')
r = [-1,1]
X, Y = np.meshgrid(r, r)
verts = [[Z[0],Z[1],Z[2],Z[3]],
 [Z[4],Z[5],Z[6],Z[7]],
 [Z[0],Z[1],Z[5],Z[4]],
 [Z[2],Z[3],Z[7],Z[6]],
 [Z[1],Z[2],Z[6],Z[5]],
 [Z[4],Z[7],Z[3],Z[0]]]
ax.add_collection3d(Poly3DCollection(verts, facecolors=None, linewidths=0.5, edgecolors='b', alpha=0.1, linestyle='--'))
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
ax.axes.zaxis.set_ticklabels([])
for line in ax.xaxis.get_ticklines():
    line.set_visible(False)
for line in ax.yaxis.get_ticklines():
    line.set_visible(False)
for line in ax.zaxis.get_ticklines():
    line.set_visible(False)
ax.set_zlim(-1,1)
ax.set_axis_off()
gridpoints = np.linspace(-0.8,0.8,4)
nvs = np.array([(np.random.uniform(-0.8,0.8),np.random.uniform(-0.8,0.8), np.random.uniform(-0.1,0.1)) for i in range(100)])
ax.scatter(nvs[:,0],nvs[:,1],nvs[:,2]-0.05, label='NV-centre', c='r', alpha=1, s=10)
patch_1 = np.array([[-0.8, -0.8, -zheight],
                  [-0.2, -0.8, -zheight ],
                  [-0.2, -0.2, -zheight],
                  [-0.8,-0.2, -zheight],
                  [-0.8, -0.8, zheight],
                  [-0.2, -0.8, zheight ],
                  [-0.2, -0.2, zheight],
                  [-0.8, -0.2, zheight]])

pos1 = -0.1
pos2 = 0.5
offset = 0.7
patch_2 = np.array([[pos1, pos1-offset, -zheight],
                  [pos2, pos1-offset, -zheight ],
                  [pos2, pos2-offset, -zheight],
                  [pos1,pos2-offset, -zheight],
                  [pos1, pos1-offset, zheight],
                  [pos2, pos1-offset, zheight ],
                  [pos2, pos2-offset, zheight],
                  [pos1, pos2-offset, zheight]])
patch_1_centre = np.mean(patch_1,axis=0)
patch_2_centre = np.mean(patch_2,axis=0)
centres = np.vstack((patch_1_centre,patch_2_centre))
ax.scatter(centres[:,0],centres[:,1],centres[:,2], s=400, label='Logical qubit', alpha=0.95)

x,y,z = centres[0,:]

arrow = Arrow3D([x+0.018, x+0.018],[y-0.2,y+0.2],[z-0.4,z+0.5], arrowstyle='-|>', color='k', mutation_scale=30, linewidth=3)
ax.add_artist(arrow)
# ax.plot([0,0],[0,0],'r--',label='Region where physical qubits are degenerate')
x,y,z = centres[1,:]
2
arrow = Arrow3D([x+0.018, x+0.018],[y-0.2,y+0.2],[z-0.4,z+0.5], arrowstyle='-|>', color='k', mutation_scale=30, linewidth=3)
ax.add_artist(arrow)

r= 0.3
u, v = np.mgrid[0:2*np.pi:30j, 0:np.pi:15j]
x = r*np.cos(u)*np.sin(v)
y = r*np.sin(u)*np.sin(v)
z = r*np.cos(v)
ax.plot_wireframe(x-0.5, y-0.5, z, color="g",linestyle='--',label='Region where physical qubits are degenerate', linewidth=0.3)
ax.plot_wireframe(x+0.2, y-0.5, z, color="g",linestyle='--', linewidth=0.3)


for (x,y,z) in nvs:
    arrow = Arrow3D([x+0.005, x+0.005],[y-0.02,y+0.02],[z-0.1,z+0.05], arrowstyle='-|>', color='k', **arrow_prop_dict)
    ax.add_artist(arrow)


nv1, nv2 = (gridpoints[1],gridpoints[0],0), (gridpoints[2],gridpoints[0],0)
ax.xaxis.set_major_locator(plt.NullLocator())
ax.yaxis.set_major_locator(plt.NullLocator())
ax.zaxis.set_major_locator(plt.NullLocator())


# azx.add_collection3d(Poly3DCollection(verts, facecolors=None, linewidths=0.5, edgecolors='r', alpha=0, linestyle='--'))  
ax.legend(loc=4)

plt.savefig('./patches.pdf', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

point  = np.array([1, 2, 3])
normal = np.array([17, 13, 0])

# a plane is a*x+b*y+c*z+d=0
# [a,b,c] is the normal. Thus, we have to calculate
# d and we're set
d = -point.dot(normal)
numx, numy = 6,6
# create x,y
xx, yy = np.meshgrid(range(numx), range(numy))

# calculate corresponding z
z = (-normal[0] * xx - normal[1] * yy)

# plot the surface
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111, projection='3d', proj_type='ortho')
surf = ax.plot_surface(xx, yy, z, alpha=0.2, label='Magnetic Field Strength')
surf._facecolors2d = surf._facecolor3d
surf._edgecolors2d = surf._edgecolor3d


#and i would like to plot this point : 
ax.scatter(xx , yy , z,  color='green', label='Individual N-V Centres')

isolate_x = lambda value : - value / normal[1]
magfield = lambda x, y : (-normal[0] * x - normal[1] * y)
isolate_y = lambda value : - (value + (numx-1) * normal[1]) / normal[0]

for i in range(numx):
    for j in range(numy-1):
        value = magfield(i,j)
        point1 = [i,j, value]
        yval = isolate_x(value)
        if yval > numx-1:
            point2 = [isolate_y(value), numx-1, value]
        else:
            point2 = [0, isolate_x(value), value]
        ax.plot([point1[0], point2[0]], [point1[1], point2[1]], [point1[2], point2[2]],color = 'r', linestyle = '--', linewidth=0.5)
ax.plot([0,0],[0,0],[0,0],color='r',linestyle='--', label='Line of equal value')
ax.legend()
ax.set_xlabel("Qubit x-index number (arbitrary)")
ax.set_ylabel("Qubit y-index number (arbitrary)")
plt.tight_layout()

plt.savefig("./planarmagfield.pdf")
plt.show()
```
